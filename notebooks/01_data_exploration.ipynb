## California Housing â€“ Exploratory Data Analysis (EDA)


```python
import numpy as np
import pandas as pd

from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing(as_frame=True)
df = housing.frame.copy()

df.head()
```

```python
print("Shape:", df.shape)
```

```python
df.info()
```

```python
df.describe().T
```

```python
print("Target name:", housing.target_names)
print("Feature names:", housing.feature_names)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df["MedHouseVal"], bins=50)
plt.xlabel("Median House Value")
plt.ylabel("Frequency")
plt.title("Distribution of Median House Value")
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.boxplot(df["MedHouseVal"], vert=False)
plt.xlabel("Median House Value")
plt.title("Boxplot of Median House Value")
plt.show()
```

```python
skew_kurt_df = pd.DataFrame({
    "Skewness": df.skew(),
    "Kurtosis": df.kurtosis()
})

skew_kurt_df
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["MedInc"], df["MedHouseVal"], alpha=0.3)
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.title("Median Income vs House Value")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Population"], df["MedHouseVal"], alpha=0.3)
plt.xlabel("Population")
plt.ylabel("Median House Value")
plt.title("Population vs House Value")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Latitude"], df["MedHouseVal"], alpha=0.3)
plt.xlabel("Latitude")
plt.ylabel("Median House Value")
plt.title("Latitude vs House Value")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Longitude"], df["MedHouseVal"], alpha=0.3)
plt.xlabel("Longitude")
plt.ylabel("Median House Value")
plt.title("Longitude vs House Value")
plt.show()
```
